```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
filename = "pima-indians-diabetes.csv"
name = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(filename, names=name)
print(df.columns)
description = df.describe()
skin_counts = df.groupby('skin').size()
skin_counts_filtered = skin_counts[1:-2]
skin_counts_filtered
```

```python
#plotting the age distribution
skin_counts_filtered.plot()
plt.xlabel('Skin Type')
plt.ylabel('Count')
plt.title("Distribution of Skin Types")
plt.show()
```

```python
plt.hist(skin_counts_filtered, bins=10, density=True, alpha = 0.3, color='g')
plt.show()
```
